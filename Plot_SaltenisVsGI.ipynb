### Import all the relevant libraries

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
plt.style.use('ggplot')
```

```python
ls = ['-',':','-.','--',(0, (3, 5, 1, 5, 1, 5))]
mk = ['s','o','v','+','*']

functions = ['A1','A2','B1','B2','B3','C1','C2']

p = 13

k = 6
```

```python
for f in functions:
    Saltenis = pd.read_excel('Saltenis_asym.xlsx',sheet_name=f,index_col=0).loc[[(k+1)*2**s for s in range(1,p+1)]].loc[(k+1)*2**2:]
    GI = pd.read_excel('G&I.xlsx',sheet_name=f,index_col=0).loc[[(k+1)*2**(s+1) for s in range(1,p+1)]].loc[(k+1)*2**2:(k+1)*2**p]
    
    dataSets = [Saltenis,GI]

    dSnames = ['Saltenis','GlenIsaacs']
    
    dS_dic = dict(zip(dSnames,dataSets))
    
    co = -1
    for dk, dv in dS_dic.items():
        co +=1
        x_vals = dv.index
        y1 = dv.values
        plt.loglog(x_vals, y1, c='k', marker = mk[co],linestyle=ls[co],label = dk)
    plt.xlabel('Cost')
    plt.ylabel('MAE')
    plt.title(f)
    plt.legend()
    plt.show()
```
