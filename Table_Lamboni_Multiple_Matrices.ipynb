## Comparison of the multiple-matrices based Lamboni and Saltenis estimators

Let us start off by importing the relevant libraries.

```python
import numpy as np
import pandas as pd
from pandas import ExcelWriter
import tabulate
```

```python
k = 6

a2 = np.array([0,0.5,3,9,99,99])
b3 = np.array([6.42,6.42,6.42,6.42,6.42,6.42])

def A1(sm):
    return pd.Series([np.prod(sm.iloc[:,:j+1],axis=1)*(-1)**(j+1) for j in range(k)]).sum()

def A2(sm):
    return pd.Series([(np.abs(4*sm[j]-2)+a2[j])/(1+a2[j]) for j in range(k)]).product()

def A2b(sm,sn):
    return pd.Series([(np.abs(4*(sm[j]+sn[j]-np.modf(sm[j]+sn[j])[1])-2)+a2[j])/(1+a2[j]) for j in range(k)]).product()

def B1(sm):
    return pd.Series([(k-sm[j])/(k-0.5) for j in range(k)]).product()
        
def B2(sm):
    return ((1+1/k)**k)*pd.Series([sm[j]**(1/k) for j in range(k)]).product()
        
def B3(sm):
    return pd.Series([(np.abs(4*sm[j]-2)+b3[j])/(1+b3[j]) for j in range(k)]).product()
        
def C1(sm):
    return pd.Series([np.abs(4*sm[j]-2) for j in range(k)]).product()
        
def C2(sm):
    return sm.product(axis=1)*2**k

functions = [A1, A2, B1, B2, B3, C1, C2]
```

### Let us now import the databases we will be working on

```python
p = 13

mm_l = []
l_l = []

for f in functions:
    Matrices2 = pd.read_excel('Saltenis_sym.xlsx',sheet_name=f.__name__,index_col=0).iloc[[2**s-1 for s in range(2,p+1)]].T
    Matrices2.columns = [2**s for s in range(2,p+1)]
    Matrices2['matrices']=2
    
    Matrices3 = pd.read_excel('3Multiple_matrices.xlsx',sheet_name=f.__name__,index_col=0).iloc[[2**s-1 for s in range(2,p+1)]].T
    Matrices3.columns = [2**s for s in range(2,p+1)]
    Matrices3['matrices']=3
    
    Matrices4 = pd.read_excel('4Multiple_matrices.xlsx',sheet_name=f.__name__,index_col=0).iloc[[2**s-1 for s in range(2,p+1)]].T
    Matrices4.columns = [2**s for s in range(2,p+1)]
    Matrices4['matrices']=4
    
    Matrices6 = pd.read_excel('6Multiple_matrices.xlsx',sheet_name=f.__name__,index_col=0).iloc[[2**s-1 for s in range(2,p+1)]].T
    Matrices6.columns = [2**s for s in range(2,p+1)]
    Matrices6['matrices']=6
    
    MultMatrices_f = pd.concat([Matrices2,Matrices3,Matrices4,Matrices6])
    MultMatrices_f['function']=f.__name__
    MultMatrices_f = MultMatrices_f.set_index(['function','matrices'])
    mm_l.append(MultMatrices_f)
    
    Lamboni2 = pd.read_excel('2Lamboni.xlsx',sheet_name=f.__name__,index_col=0).iloc[[2**s-1 for s in range(2,p+1)]].T
    Lamboni2.columns = [2**s for s in range(2,p+1)]
    Lamboni2['matrices']=2
    
    Lamboni3 = pd.read_excel('3Lamboni.xlsx',sheet_name=f.__name__,index_col=0).iloc[[2**s-1 for s in range(2,p+1)]].T
    Lamboni3.columns = [2**s for s in range(2,p+1)]
    Lamboni3['matrices']=3
    
    Lamboni4 = pd.read_excel('4Lamboni.xlsx',sheet_name=f.__name__,index_col=0).iloc[[2**s-1 for s in range(2,p+1)]].T
    Lamboni4.columns = [2**s for s in range(2,p+1)]
    Lamboni4['matrices']=4
    
    Lamboni6 = pd.read_excel('6Lamboni.xlsx',sheet_name=f.__name__,index_col=0).iloc[[2**s-1 for s in range(2,p+1)]].T
    Lamboni6.columns = [2**s for s in range(2,p+1)]
    Lamboni6['matrices']=6
    
    Lamboni_f = pd.concat([Lamboni2,Lamboni3,Lamboni4,Lamboni6])
    Lamboni_f['function']=f.__name__
    Lamboni_f = Lamboni_f.set_index(['function','matrices'])
    
    l_l.append(Lamboni_f)
    
MultMatrices = pd.concat(mm_l)
Lamboni = pd.concat(l_l)

Percentage_Error = ((MultMatrices-Lamboni)*100/MultMatrices).astype(int)
Percentage_Error
```
